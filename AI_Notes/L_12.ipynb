# **1. Introduction**

**Beam Search** is an **informed (heuristic) search algorithm** and a **restricted version of Best-First Search**.
It is mainly designed to solve the **memory explosion problem** in large search spaces by keeping only a *fixed number of best nodes* (called **beam width = K**) at each level.

* **Type of Search:** Heuristic / Informed search
* **Parent Algorithm:** Greedy Best-First Search
* **Key Feature:** Prunes aggressively to save memory
* **Major Trade-off:** Saves space, but loses completeness & optimality

---

# **2. Why Beam Search? (Motivation)**

Standard Best-First Search expands nodes based on heuristic values but **stores all generated nodes** in memory.

This leads to:

* **Space Complexity = O(báµˆ)** â†’ exponential
* Memory blow-up in deep/wide trees

Beam Search fixes this by **restricting the number of nodes stored at each level to a constant K**.

---

# **3. Key Concept: Beam Width (K)**

**Beam Width (K)** = the fixed number of *best* successor nodes kept at every level.

### **How it Works**

At each level:

1. Generate all successor nodes
2. Compute heuristic value `h(n)` for each
3. **Select only top K nodes (lowest h)**
4. **Prune all other nodes permanently**

This makes the search lightweight, fast, and memory-friendly.

---

# **4. Algorithmic Properties**

### âœ” **Advantages**

* **Space-efficient** â†’ only K nodes stored, not thousands
* **Faster than Best-First Search**
* Avoids exploring unpromising branches

### âŒ **Disadvantages**

* **Not complete**

  * If the optimal path is pruned early, solution is lost
* **Not optimal**

  * Works greedily based only on local heuristic decisions

### **Q1. Differentiate between Best-First Search and Beam Search.**

**Answer:**
Both Best-First Search and Beam Search are heuristic-based informed search techniques. Best-First Search expands nodes by selecting the node with the lowest heuristic value from the entire open list. Its main drawback is **exponential space complexity (O(báµˆ))**, as it stores all generated nodes.

Beam Search solves this limitation by introducing a **beam width (K)**, which restricts the number of nodes retained for future expansion. At every level, only the best K nodes (with lowest heuristic values) are kept, and all others are pruned. This reduces space complexity to **O(KÂ·d)**. However, due to pruning, Beam Search becomes **incomplete and non-optimal**.

---

### **Q2. Explain how beam width (K) affects the performance of Beam Search.**

**Answer:**
Beam width (K) is the most important parameter in Beam Search. It determines how many nodes are kept at each level.

* **Small K (e.g., K = 1):**

  * Very fast, very low memory use
  * But extremely risky
  * High chance of missing the optimal path â†’ incomplete & non-optimal

* **Large K (e.g., K â†’ âˆž):**

  * Approaches Best-First Search
  * Higher chance of finding complete & optimal solution
  * But memory and time costs increase

Thus, K must be chosen carefully to balance **speed, memory, and solution quality**.
