# **5. Challenges & Need for Better Agents**

The Simple Reflex Agent has multiple weaknesses:

### **1. Non-Optimal Decisions**

Acts only based on current percept, so it may:

* Turn on AC even if nobody is in the room
* Move left/right blindly without planning

### **2. Rule Explosion**

Complex environments â†’ too many Ifâ€“Then rules â†’ impossible to manage.

### **3. Fails in Partial Observability**

If sensors cannot observe full environment, the agent:

* Gets confused
* Takes wrong actions
* Cannot infer hidden information

This is why more advanced agents (Model-Based, Goal-Based, Utility-Based) are needed.

## **Q1. Define the Simple Reflex Agent and explain its fundamental limitation in partially observable environments.**

### **Answer:**

A Simple Reflex Agent is an AI agent that chooses its actions based solely on the **current percept**, using predefined **Conditionâ€“Action rules** (Ifâ€“Then rules). It does not store percept history or maintain internal memory.

Its major limitation in **partially observable** environments is the lack of memory. If sensors cannot observe crucial information, the agent has **no past percepts to infer the true state**. This leads to incorrect or non-optimal actions because it cannot reason beyond the immediate input.
