## **1. Introduction to AI**

**Definition:**
Artificial Intelligence (AI) is the field of computer science focused on creating machines that can **think, behave, and work like humans**.

**Core Goal:**
Build intelligent systems capable of performing tasks that normally require human intelligence.

**Why it's important:**
AI powers everyday tech like virtual assistants, recommendation systems, autonomous cars, and smart decision-making tools.

---

* The idea of intelligent machines began around **World War II** when the first computers were being developed.

## **3. Human-Like Abilities AI Tries to Copy**

### **1. Reasoning**

* Ability to think logically and draw conclusions.
  *Example:* If pass marks = 40% and student gets 60%, AI should conclude ‚Äústudent passed.‚Äù

### **2. Learning**

* Gaining knowledge from experience and improving performance.
  *Example:* AI models improving accuracy as they are trained with more data.

### **3. Problem-Solving**

* Identifying a problem and finding solutions.
  *Example:* Game-playing AI (Chess, Sudoku).

### **4. Perception**

* Understanding the world through sensors.
  *Example:* Self-driving car detecting traffic and rerouting.

---

## **4. Major Real-World Applications of AI**

### **1. Virtual Assistants (Siri, Google Assistant)**

* Use **Natural Language Processing (NLP)** to understand voice commands and respond like humans.

### **2. Self-Driving Cars (Tesla)**

* Use sensors + AI to detect obstacles, read traffic, and drive without human input.

### **3. Recommendation Systems (YouTube, Netflix, Amazon)**

* Analyze user behavior to recommend movies, videos, or products.

### **4. Expert Systems**

* Provide expert-level decisions using user data.
  *Example:* System warns a user with high cholesterol when ordering unhealthy food.

---

## **5. Natural Language Processing (NLP)**

**Definition:**
NLP is a branch of AI that helps computers understand and process human language (speech or text).

**Uses:**

* Voice assistants
* Translators
* Chatbots
* Sentiment analysis
