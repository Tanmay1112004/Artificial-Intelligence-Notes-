# **1. Introduction to Intelligent Agents**

An **Intelligent Agent** is any entity (human, robot, software, machine) that **perceives** its environment and **acts** upon it to achieve goals.

### **Agent Cycle (Core Idea)**

Agents work in a constant feedback loop:

1. **Perception (Sense):**
   Uses **Sensors** to collect information (percepts) from the environment.

2. **Decision-Making:**
   The **Agent Program** processes percepts ‚Üí decides the best action.

3. **Action (Actuate):**
   Uses **Actuators** to perform actions that affect the environment.

This cycle repeats continuously.

---

# **2. Components of an Intelligent Agent**

### **1. Sensors (Input Devices)**

Used to observe the environment.

* **Human Examples:** Eyes, ears, nose, skin
* **Robot Examples:** Camera, microphone, GPS, IR sensor

### **2. Actuators / Effectors (Output Devices)**

Used to take action.

* **Human Examples:** Hands, legs, vocal cords
* **Robot Examples:** Wheels, arms, display, speakers

### **3. Agent Program (The Brain)**

* A function that maps **percept sequences ‚Üí actions**.
* Defines how the agent thinks and behaves.

---

# **3. Rationality and Performance**

### **Goal of AI Agents:**

To perform **rational actions** and achieve **high performance**.

### **Rational Action (Definition)**

An action is **rational** if it:

* Maximizes the performance measure
* Uses all available percepts, knowledge, history
* Is the best choice based on what the agent knows

### **Performance Measure**

Criteria for judging the agent‚Äôs success.
Examples: accuracy, speed, safety, efficiency.

---

# **4. PEAS Framework**

Used for designing and analyzing any intelligent agent.

| Component | Meaning             | Example: Self-Driving Car                     |
| --------- | ------------------- | --------------------------------------------- |
| **P**     | Performance Measure | Safety, speed, comfort, time, fuel efficiency |
| **E**     | Environment         | Roads, pedestrians, traffic, weather          |
| **A**     | Actuators           | Steering, brakes, accelerator, gear           |
| **S**     | Sensors             | Cameras, GPS, radar, LIDAR, speedometer       |

# **5. Types of Intelligent Agents**

(Just short definitions ‚Äî exactly what examiners want)

1. **Simple Reflex Agents**

   * Act only on *current* percept.
   * No memory.

2. **Model-Based Reflex Agents**

   * Maintain internal state (memory).
   * Use history + current percept.

3. **Goal-Based Agents**

   * Choose actions based on goals.

4. **Utility-Based Agents**

   * Maximize utility (happiness/satisfaction level)

5. **Learning Agents**

   * Learn from past experience.
   * Improve performance over time.

## **Q1. Describe the continuous cycle of operation for an Intelligent Agent.**

**Answer:**
An Intelligent Agent works in a continuous loop consisting of:

1. **Perception:**
   Sensors capture percepts (current + historical data) from the environment.

2. **Decision-Making:**
   The Agent Program uses these percepts to decide the best action.
   It maps percept sequences to actions.

3. **Action:**
   Actuators execute the chosen action and modify the environment.

After acting, the agent senses the new state and repeats the cycle indefinitely.

---

## **Q2. What is a Rational Action? Why is it important in AI design?**

**Answer:**
A **Rational Action** is the action that is expected to **maximize an agent‚Äôs performance measure**, based on its percept history and knowledge.

### **Importance:**

* Ensures the agent moves toward its goal effectively.
* Helps the agent choose the best possible action in each situation.
* Directly improves performance and decision-making.
* Forms the foundation of designing high-performance and optimal AI systems.
